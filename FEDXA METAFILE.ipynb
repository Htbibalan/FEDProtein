```python
import trompy as tp
import numpy as np
import csv
%matplotlib inline
from datetime import datetime
```

```python
def get_FEDevents (filename, eventname):
    file= open (filename)
    csvreader= csv.reader(file)
    header=[]
    header= next(csvreader)
    rows =[]
    for row in csvreader:
        row.append(row)
    timestamps=[]
    for row in rows:
        if row[7]== eventname:
            date_time_obj = datetime.strptime(row [0],'%m/%d/%Y %H:%M:%S')
            timestamps.append(date_time_obj)
            
    t0= [0]
    day = t0.split()[0]
    lightson = day + " 07:00:00"
    refpoint= datetime.strptime(lightson,'%m/%d/%Y %H:%M:%S' )
    
    pellettimes= []
    for t in timestamps:
        Deltat= t-refpoint
        Deltainseconds = Deltat.total_seconds()
        Deltatinhours = Deltainseconds/3600
        pellettimes.append(Deltainhours)
    return pellettimes
        
```
